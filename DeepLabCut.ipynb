## **DeepLabCut Instructions**

Launch the Terminal:

```python
conda --version
```

If it says more than 24, move forward. Here, the next step is to **create a virtual environment**

```python
conda create -n deeplabcut python=3.10
```

Say "y" x2

```python
conda activate deeplabcut
```

**Installing DeepLabCut GUI**

```python
pip install "deeplabcut[gui]"
```

```python
python -m deeplabcut
```

After doing the "python -m deeplabcut," the terminal will need some time. Then, it will launch DeepLabCut

**Starting a Project**

1. Select Create Project and give it a name
2. In the bodyparts track, you will add the following:
- left_antena
- right_antena
- head
- abdomen

3. Click "select individual files"
4. Click "browse videos" and upload manual each video that you need.
5. Click "create" on the bottom right corner

Follow the bar in order of left to right to complete the entire process for data outputs.

**Manage Project**
1. Click on "Add New Videos" and upload all (shift can be used here)
2. When successfully uploaded, at the bottom it will tell you that new videos were uploaded.

**Extract Frames**
1. Keeping the same settings, got to "select videos" and you will only be able to choose one at a time. Get all videos.
2. Click "extract frames"
3. Loading bar at the bottom will appear (if multiple videos, it will take a while)

**Label Frames**
1. Click on "Label Frames"
2. Select one video of interest from the "Labeled data" folder
3. Once video pops up, you will need to go frame by frame and annotate the keypoint body parts.
4. First, make sure you know what is the keypoint selection you have. Then, click the add circle.

- To delete any point, go to the left bottom and click on the name of the body part and delete.
- After labelling each frame, review each one to make sure they are all correct
5. command + S to save the labelling
At the bottom of the window, it will tell you "Data successfully saved"
6. Then X out and continue with other videos

**Create Training Dataset**
1. Keeping the same settings, and click on "Create Training Dataset
2. A success message will appear and click "ok"

- Ex: If you start with 5 videos and then you want to add another 5 videos, then you will have to change "shuffle" from 1 to 2

**Train Network**
1. Keeping the same settings, then **connect to CIMeC GPU** at: https://wiki.cimec.unitn.it/tiki-index.php?page=ACNGpu or use the file on Google Collab (make sure to change the running time to T4 GPU)
2. Open a new file in VSCODE
1. Install anaconda 3

- If at any moment you need to stop, control + C

2. click "Train Network"
